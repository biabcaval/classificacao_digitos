```python
import pandas as pd
import numpy as np

df = pd.read_csv("test.csv", sep=';')
df
```

```python
# criando os datasets reduzidos
num_linhas = 1115


df_teste_redu = pd.DataFrame({
    'intensidade': [0 for i in range(num_linhas)],
    'simetria': [0 for i in range(num_linhas)],
    'label': [l for l in df['label']]
})

df_sem_label = df.loc[:, 'pixel0':'pixel783']

# intensidade
for linha in range(1115):
    i = 0
    for x in range(784):
        i+= df_sem_label[f'pixel{str(x)}'].loc[linha]
    i /= 255
    df_teste_redu['intensidade'].loc[linha] = i



```

```python
df_teste_redu
```

```python


# simetria vertical
svs = []
for linha in range(1115):
    sv = 0

    for i in range(28):
        Pv = 0
        for j in range(14):
            pixel1 = i * 28 + j # i*28 é para chegar até a linha i, j é o que ele "anda" para chegar até a coluna j
            pixel2 = i * 27 + (28 - j) # i*28 é para chegar até a linha i, j é o que ele "anda" da direita pra esquerda para chegar até a coluna
            Pv += abs(df_sem_label.iloc[linha, pixel1+1] - df_sem_label.iloc[linha, pixel2+1])
    sv = Pv/255
    svs.append(sv)

# simetria horizontal
shs = []
for linha in range(1115):
    sh = 0

    for j in range(14):
        Ph = 0
        for i in range(28):
            pixel1 = i * 28 + j
            pixel2 = (27-i) * 28 + j
            Ph += abs(df_sem_label.iloc[linha, pixel1+1] - df_sem_label.iloc[linha, pixel2+1])
    sh = Ph/255
    shs.append(sh)


simetria_total = [v + h for v, h in zip(svs, shs)]


```
